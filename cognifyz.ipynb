```python
# Here i am creating function to fill nan values of cuisines column on the basis of Locality
def fill_nan_cuisines(df):
    common_cuisines=df.groupby('Locality')['Cuisines'].apply(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
    
    df['Cuisines']=df.apply(lambda row : common_cuisines[row['Locality']] if pd.isna(row['Cuisines']) else row['Cuisines'] , axis=1)
    return df
    
```

```python
from scipy.spatial import distance
```

```python
# How to fill these nan values i think we can fill these nan values using longitude and latitude .
def fill_nan_lon_lat(df):
    # create df for nan values and without nan values
    df_nan=df[df['Cuisines'].isna()]
    df_without_nan=df[~df['Cuisines'].isna()]
    # convert df_nan and df_without_nan into numpy arrays
    cords_na=df_nan[['Longitude','Latitude']].to_numpy()
    cords_without_na=df_without_nan[['Longitude','Latitude']].to_numpy()
    
    # iterate over rows with nan cuisines
    for i, row  in df_nan.iterrows():
        # calculate distances to all nan rows
        row_coords = np.array([row['Longitude'], row['Latitude']]).reshape(1, -1)
        distances=distance.cdist(row_coords,cords_without_na, 'euclidean').flatten()
        
        # now we get distances , now find indexes of nearest distances
        nearest_index=np.argmin(distances)
        
        # now we got indexes so we can fill the cuisines column
        df.at[i, 'Cuisines'] = df_without_nan.iloc[nearest_index]['Cuisines']
    return df 
```

```python
df=fill_nan_lon_lat(df)
```

## Here we would doing some feature engneering so model perform well .

```python
# here we create count of cuisines for each resturant . so model get the additional feature 
# to predict rating on the basis of number of cuisines are available in that resturant
df['Cuisine Count'] = df['Cuisines'].str.count(',') + 1
```

```python
df['Cuisines'] = df['Cuisines'].apply(lambda x: [cuisine.strip() for cuisine in x.split(',')])
# remove leading , trailing white spaces
```

```python
# create list 
cuisine_series = df.explode('Cuisines')
```

```python
# calculate each cuisine rating
cuisine_mean_rating = cuisine_series.groupby('Cuisines')['Aggregate rating'].mean()
```

```python
# asign avg rating for each cuisines 
df['Cuisines_Mean_Rating'] = df['Cuisines'].apply(lambda x: [cuisine_mean_rating[cuisine] for cuisine in x])
```

```python
# calculate mean of that cuisines 
df['Cuisines_Mean_Rating_Avg'] = df['Cuisines_Mean_Rating'].apply(lambda x: sum(x)/len(x))
```

```python
df[['Cuisines', 'Cuisines_Mean_Rating', 'Cuisines_Mean_Rating_Avg']].head()
```

```python
df['City'] = df['City'].str.replace('Bras_lia', 'Braslia').str.replace('So Paulo', 'So Paulo')
```

```python
df[0:5]
```

```python
city_mean_rating = df[df['Aggregate rating'] != 0.0].groupby('City')['Aggregate rating'].mean()
```

```python
def replace_with_city_mean(row):
    if row['Aggregate rating'] == 0.0:
        return city_mean_rating.get(row['City'], row['Aggregate rating'])
    return row['Aggregate rating']
df['Aggregate rating'] = df.apply(replace_with_city_mean, axis=1)
```

```python
# here we are doing encoding 
df['Is delivering now'] = df['Is delivering now'].map({'Yes': 1, 'No': 0})
```

```python
df[0:5]
```

```python
# here we are doing encoding 
df['Switch to order menu'] = df['Switch to order menu'].map({'Yes': 1, 'No': 0})
```

```python
# here  we created price table booking column . more expensive restaurants might be expected to offer table booking, and 
# if they do, this might correlate with higher ratings. Conversely, 
# if a high-priced restaurant does not offer table booking, this might negatively impact its ratings. so this feature will helps model 
df['Price_TableBooking'] = df['Price range'] * df['Has Table booking']
```

```python
# here we would doing ordinal encoding of rating color and rating text
```

```python
df.groupby(['Rating color','Rating text'])['Aggregate rating'].mean()
```

```python
ordinal_mapping_for_color={'Dark Green' : 5,'Green' : 4,'Orange':2,'Red':1,'White': 0,'Yellow':3}
df['Rating color']=df['Rating color'].map(ordinal_mapping_for_color)
```

```python
ordinal_mapping = {
    'Poor': 1,
    'Average': 2,
    'Good': 3,
    'Very Good': 4,
    'Excellent': 5,
    'Not rated': 0}

df['Rating text']=df['Rating text'].map(ordinal_mapping)

```

```python
# here we would calculate avg_cost_for_one in inr format

def convert_to_inr(row):
     
    exchange_rates = {
        'Botswana Pula(P)': 6.3,     
        'Brazilian Real(R$)': 22.32,  
        'Dollar($)': 83.78,            
        'Emirati Diram(AED)': 22.81, 
        'Indonesian Rupiah(IDR)': 0.0054, 
        'NewZealand($)': 51.33,         
        'Pounds()': 108.86,            
        'Qatari Rial(QR)': 23.08,     
        'Rand(R)': 4.72,              
        'Sri Lankan Rupee(LKR)': 0.28, 
        'Turkish Lira(TL)':2.48     
        
    }
    
    currency = row['Currency']
    avg_cost_for_one = row['avg_cost_for_one']
    
    if currency == 'Indian Rupees(Rs.)':
        return avg_cost_for_one
    else:
        
        avg_cost_for_one = avg_cost_for_one * exchange_rates.get(currency, 1)
        return avg_cost_for_one
```

```python
df['avg_cost_for_one']=df.apply(convert_to_inr, axis=1)
```
